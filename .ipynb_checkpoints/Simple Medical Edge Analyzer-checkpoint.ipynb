# Simple Medical Edge Analyzer

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

## Display Image

```python
img = cv2.imread('Images/chest_xray.JPG')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img);
```

## Detect Edges

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray');
```

```python
blur = cv2.GaussianBlur(gray,(5,5),0)
plt.imshow(blur,cmap='gray');
```

```python
edges = cv2.Canny(blur,50,150)
plt.imshow(edges);
```

```python
filter = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,3)
plt.imshow(filter,cmap='gray');
```

```python
med_val = np.median(img)
```

```python
lower = int(max(0,0.2*med_val))
upper = int(min(0,1.1*med_val))
```

```python
blurred_image = cv2.blur(img,ksize=(5,5))
```

```python
edges= cv2.Canny(blurred_image ,lower,upper)
plt.imshow(edges);
```

```python
edges = cv2.Canny(image=filter,threshold1=150,threshold2=255)
plt.imshow(edges);
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
enhanced = clahe.apply(gray)
```

```python
blurred = blur = cv2.GaussianBlur(gray,(5,5),0)
```

```python
edges = cv2.Canny(blurred, 30,90)
```

```python
kernel = np.ones((2,2),np.uint8)
edges_clean = cv2.morphologyEx(edges, cv2.MORPH_CLOSE,kernel)
```

```python
img_color = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)
img_color[edges_clean == 255] = [0,255,0]
```
